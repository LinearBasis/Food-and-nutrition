```python
def get_title(list_of_ingredients, df):
	for i in list_of_ingredients:
		if not i in df.columns.to_list():
			print('Wrong ingridient(s)')
			return
	condition = 1
	for ingredient in list_of_ingredients:
		condition &= (new_df[ingredient] == 1)
	dishes = new_df[condition].sort_values(['rating'], ascending = False)[['title', 'rating']]
	dishes = dishes.iloc[:3]
	return dishes
```

```python
get_title(['apple', 'turkey'], new_df)
```

```python
list_of_ingredients = ['apple', 'turkey']
condition = 1
for ingredient in list_of_ingredients:
    condition &= (new_df[ingredient] == 1)
dishes = new_df[condition].sort_values(['rating'], ascending = False)['title']
dishes = dishes.iloc[:3]
dishes = [dish.strip().replace(' ', '-') for dish in dishes]
dishes
```

```python
def get_url(dishes):
	sep_dishes = dishes
	sep_dishes.strip().replace(' ', '-')
	search_url = f'https://www.epicurious.com/search/{sep_dishes}'
	search_resp = requests.get(search_url, headers={'User-Agent': 'Custom'})
	bs = BeautifulSoup(search_resp.text, 'html.parser')
	urls = bs.find_all('a', href=True)
	refs = bs.find()
	for url in urls:
		if url.get_text() == dishes:
			return 'https://www.epicurious.com' + url.get('href')
```

```python
def recommendation(list_of_ingredients, df):
    print("III. TOP-3 SIMILAR RECIPES:")
    title_rating = get_title(list_of_ingredients, df)
    #print(title_rating)
    for i, row in title_rating.iterrows():
        dish = row.title.strip()
        rating = row.rating
        url = get_url(dish)
        print(f'{dish}, rating: {rating}, URL: {url}')
        
```
